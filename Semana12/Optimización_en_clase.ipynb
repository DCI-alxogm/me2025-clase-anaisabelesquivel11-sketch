```python
jac_fx= lambda x: -2*np.cos(x)+x/5
hess_fx= lambda x:  2*np.sin(x)+1/5
minimize(fx,x0=0, jac=jac_fx, hess=hess_fx,  method='Newton-CG', tol=1e-8)
```
